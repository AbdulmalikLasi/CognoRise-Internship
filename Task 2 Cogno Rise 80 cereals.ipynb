• fat: grams of fat

• sodium: milligrams of sodium

• fiber: grams of dietary fiber

• carbo: grams of complex carbohydrates

• sugars: grams of sugars

• potass: milligrams of potassium

• vitamins: vitamins and minerals - 0, 25, or 100, indicating the typical percentage of FDA recommended

• shelf: display shelf (1, 2, or 3, counting from the floor)

• weight: weight in ounces of one serving

• cups: number of cups in one serving

• rating: a rating of the cereals Possibly from Consumer Reports?

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
cereals=pd.read_csv(r"C:\Users\Abdulmalik Lasisi\Downloads\cereal.csv")
```

```python
cereals.info()
```

```python
cereals['mfr'] = cereals['mfr'].replace({ 'A':'American Home Food Products','G':'General Mills','K':'Kelloggs','N':'Nabisco','P':'Post','Q':'Quaker Oats','R':'Ralston Purina'})

cereals['type'] = cereals['type'].replace({'C':'Cold','H':'Hot'})
```

```python
cereals.head()
```

```python
cereals["shelf"]= cereals["shelf"].astype(object)
cereals["vitamins"]=cereals["vitamins"].astype(object)
cereals["protein"]=cereals["protein"].astype(object)
cereals["fat"]=cereals["fat"].astype(object)
```

```python
cereals["fat"].value_counts()
```

```python
def discrete_univariate_analysis(discrete_data):
    for col_name in discrete_data:
        print("*"*10, col_name, "*"*10)
        print(discrete_data[col_name].agg(['count', 'nunique', 'unique']))
        print('Value Counts: \n', discrete_data[col_name].value_counts())
        print()
```

```python
discrete_df = cereals.select_dtypes(include=['object'])
discrete_univariate_analysis(discrete_df)
```

```python
numeric_cereals = cereals.select_dtypes(include=[np.number])

# Compute the correlation matrix
corr = numeric_cereals.corr()

# Plot the correlation matrix as a heatmap
plt.figure(figsize=(12, 12), dpi = 100)
sns.heatmap(corr, annot=True, fmt=".2f", cmap="cool", annot_kws={"size": 10})
plt.show()
```

```python
fig,axs = plt.subplots(figsize=(13,3),layout="constrained")
sns.countplot(data=cereals,x='mfr',ax=axs)
```

```python
fig,axs = plt.subplots(figsize=(5,3),layout="constrained")
sns.countplot(data=cereals,x='type',ax=axs)
```

```python
fig,axs=plt.subplots(1,2,figsize=(9,7),layout="constrained")
p=cereals["mfr"].value_counts()
axs[0].pie(p.values,labels=p.index,autopct="%1.1f%%")
q=cereals["type"].value_counts()
axs[1].pie(q.values,labels=q.index,autopct="%1.1f%%")
plt.show()
```
